```python
class State:
    
    ###########################################################################
    #Initialization of Eight Puzzle by the given fringe, size and the visited list
    # fringe = List of expanded states
    # visited = List of visited states
    ###########################################################################
    
    def __init__(self,val,lev,fval):
        self.val = val
        self.lev = lev
        self.fval = fval
        
    ##########################################################################################################  
    # This function will generate the children of the parent by moving the zero-(0) elememnt in the feasible postion
    # There are 4 possible moves : left,up,right,down
    # Feasible position depends on where the element zero(0) is placed.
    # If 0 is in corner, then 2 possible moves.
    # If 0 is at center, then 4 possible moves.
    # If 0 is an edge, then 3 possible moves.
    ##########################################################################################################
    
    def generate_children(self):

        x,y = self.search_space(self.val,'0')
        coordination_list = [[x,y-1],[x,y+1],[x-1,y],[x+1,y]]
        children = []
        for i in coordination_list:
            child = self.valid_position(self.val,x,y,i[0],i[1])
            if child is not None:
                child_node = State(child,self.lev+1,0)
                children.append(child_node)
                
        return children
      
    ######################################################################################
    # This Function find the position of the 0
    ######################################################################################
    
    def search_space(self,puzzle,x):
        for i in range(0,len(self.val)):
            for j in range(0,len(self.val)):
                if puzzle[i][j] == x:
                    return i,j
                
    ######################################################################
    # The function will Check for the valid position of the 0 and returns it.
    ######################################################################
    
    def valid_position(self,puzzle,x1,y1,x2,y2):
        
        if x2 >= 0 and x2 < len(self.val) and y2 >= 0 and y2 < len(self.val):
            temp_puzzle = []
            temp_puzzle = self.c_matrix(puzzle)
            temp = temp_puzzle[x2][y2]
            temp_puzzle[x2][y2] = temp_puzzle[x1][y1]
            temp_puzzle[x1][y1] = temp
            return temp_puzzle
        else:
            return None
        
    ##########################    
    # Makes the copy of matrix
    ##########################
    
    def c_matrix(self,root):

        temp_arr = []
        for i in root:
            t = []
            for j in i:
                t.append(j)
            temp_arr.append(t)
        return temp_arr    
            
   
```
